```python
from scipy.stats import norminvgauss, laplace, poisson, cauchy, uniform
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

Блок заведения переменных

```python
sizes = [10, 100, 1000]
NORMAL, CAUCHY, LAPLACE, POISSON, UNIFORM = "x", "x", "x", "x", "x"
N, C, L, P, U = "NormGistag", "CauchyGistag", "LapGistag", "PoissonGistag", "UniGistag"
DENSITY, SIZE = "f(x)", "Size: "
LINE_TYPE = 'k--'
HISTOGRAM_TYPE = 'stepfilled'
COLORS = ["crimson", "blueviolet", "aqua", "lime", "antiquewhite"]
EXPANSION = '.png'
FILEPATH = 'graphs/'
```

Функция построения гистограмм и графиков нормального распределения в зависимости от кол-ва элементов выборки

```python
def Normal():
    for size in sizes:
        density = norminvgauss(1, 0)
        histogram = norminvgauss.rvs(1, 0, size = size)
        fig, ax = plt.subplots(1, 1)
        ax.hist(histogram, density = True, histtype = HISTOGRAM_TYPE, alpha = 0.6, color = COLORS[0])
        x = np.linspace(density.ppf(0.01), density.ppf(0.99), 100)
        ax.plot(x, density.pdf(x), LINE_TYPE, lw = 1.5)
        ax.set_xlabel(NORMAL)
        ax.set_ylabel(DENSITY)
        ax.set_title(SIZE + str(size))
        plt.grid()
        plt.show()
        fig.savefig(str(FILEPATH) + N + str(size) + EXPANSION)
    return

Normal()
```

Функция построения гистограмм и графиков распределения Коши в зависимости от кол-ва элементов выборки

```python
def Cauchy():
    for size in sizes:
        density = cauchy()
        histogram = cauchy.rvs(size=size)
        fig, ax = plt.subplots(1, 1)
        ax.hist(histogram, density=True, histtype=HISTOGRAM_TYPE, alpha=0.6, color=COLORS[1])
        x = np.linspace(density.ppf(0.01), density.ppf(0.99), 100)
        ax.plot(x, density.pdf(x), LINE_TYPE, lw=1.5)
        ax.set_xlabel(CAUCHY)
        ax.set_ylabel(DENSITY)
        ax.set_title(SIZE + str(size))
        plt.grid()
        plt.show()
        fig.savefig(str(FILEPATH) + C
                    + str(size) + EXPANSION)
    return

Cauchy()
```

Функция построения гистограмм и графиков распределения Лапласа в зависимости от кол-ва элементов выборки

```python
def Laplace(): 
    for size in sizes:
        density = laplace(scale=1 / m.sqrt(2), loc=0)
        histogram = laplace.rvs(size=size, scale=1 / m.sqrt(2), loc=0)
        fig, ax = plt.subplots(1, 1)
        ax.hist(histogram, density=True, histtype=HISTOGRAM_TYPE, alpha=0.6, color=COLORS[2])
        x = np.linspace(density.ppf(0.01), density.ppf(0.99), 100)
        ax.plot(x, density.pdf(x), LINE_TYPE, lw=1.5)
        ax.set_xlabel(LAPLACE)
        ax.set_ylabel(DENSITY)
        ax.set_title(SIZE + str(size))
        plt.grid()
        plt.show()
        fig.savefig(str(FILEPATH) + L + str(size) + EXPANSION)
    return

Laplace()
```

Функция построения гистограмм и графиков распределения Пуассона в зависимости от количества элементов выборки

```python
def Poisson(): 
    for size in sizes:
        density = poisson(10)
        histogram = poisson.rvs(10, size=size)
        fig, ax = plt.subplots(1, 1)
        ax.hist(histogram, density=True, histtype=HISTOGRAM_TYPE, alpha=0.6, color=COLORS[3])
        x = np.arange(poisson.ppf(0.01, 10), poisson.ppf(0.99, 10))
        ax.plot(x, density.pmf(x), LINE_TYPE, lw=1.5)
        ax.set_xlabel(POISSON)
        ax.set_ylabel(DENSITY)
        ax.set_title(SIZE + str(size))
        plt.grid()
        plt.show()
        fig.savefig(str(FILEPATH) + P + str(size) + EXPANSION)
    return

Poisson()
```

Функция построения гистограмм и графиков равномерного распределения в зависимости от количества элементов выборки

```python
def Uniform(): 
    for size in sizes:
        density = uniform(loc=-m.sqrt(3), scale=2 * m.sqrt(3))
        histogram = uniform.rvs(size=size, loc=-m.sqrt(3), scale=2 * m.sqrt(3))
        fig, ax = plt.subplots(1, 1)
        ax.hist(histogram, density=True, histtype=HISTOGRAM_TYPE, alpha=1, color=COLORS[4])
        x = np.linspace(density.ppf(0.01), density.ppf(0.99), 100)
        ax.plot(x, density.pdf(x), LINE_TYPE, lw=1.5)
        ax.set_xlabel(UNIFORM)
        ax.set_ylabel(DENSITY)
        ax.set_title(SIZE + str(size))
        plt.grid()
        plt.show()
        fig.savefig(str(FILEPATH) + U + str(size) + EXPANSION)
    return

Uniform()
```
